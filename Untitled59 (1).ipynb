```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# this function to roll two dice and return the sum
def RTdice():                      # RTdice is roll two dice
    Dice1 = random.randint(1, 6)
    Dice2 = random.randint(1, 6)
    return Dice1 + Dice2

# here is the function to roll dice a specified number of times and count frequencies of sums
def RDCF(rolls):              #RDCF is the roll dice and count frequencies
    frequencies = {sum_value: 0 for sum_value in range(2, 13)} 
    for _ in range(rolls):
        roll_sum = RTdice()
        frequencies[roll_sum] += 1
    return frequencies

# function to plot the frequencies using seaborn
def plot_frequencies(frequencies, rolls):
    df = pd.DataFrame(list(frequencies.items()), columns=['Sum', 'Frequency'])
    sns.barplot(x='Frequency', y='Sum', data=df, hue='Sum', palette="muted", orient='h', legend=False)
    plt.title(f'Frequencies for {rolls} Rolls')
    plt.xlabel('Frequency')
    plt.ylabel('Dice Value')
    plt.gca().invert_yaxis()
    plt.show()


try:
    rolls = int(input("Enter the number of rolls (e.g., 360, 3600, 36000): "))
    if rolls <= 0:
        raise ValueError("Number of rolls must be a positive integer.")
except ValueError as e:
    print(f"Invalid input: {e}")
else:
    frequencies = RDCF(rolls)
    plot_frequencies(frequencies, rolls)
```
